```python
foot=get_dummies(info,columns=['Preferred Foot'],drop_first=True)
foot.head(5)
```

```python
age=get_dummies(info,columns=['Age'])
age.head()
```

```python
new_info=concat([info,foot,age],axis=1)
new_info.head(5)
new_info.drop(new_info.columns[[1,2,3,4,5]],axis=1,inplace=True)
new_info.head(5)
```

```python
from sklearn.linear_model import *
x=new_info.iloc[:,1:]
y=new_info.iloc[:,1]
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
re=LogisticRegression(solver='lbfgs',random_state=0)
re.fit(x_train,y_train)
predict=re.predict(x_test)
print(predict)
h=classification_report(y_test,predict)
print(h)
j=accuracy_score(y_test,predict)
print(j)
k=confusion_matrix(y_test,predict)
print(k)
```
